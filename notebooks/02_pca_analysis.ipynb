# PCA Analysis

## 1. Impoting Libraries

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

## 2. Load cleaned data

```python
df = pd.read_csv('../data/heart_disease_cleaned.csv')
X = df.drop('num', axis=1)
y = df['num']
```

## 3. Apply PCA

```python
pca = PCA()
X_pca = pca.fit_transform(X)
```

## 4. Plot explained variance

```python
plt.figure(figsize=(10, 6))
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('Number of Components')
plt.ylabel('Cumulative Explained Variance')
plt.title('PCA Explained Variance')
plt.grid()
plt.savefig('../results/pca_variance.png')
plt.show()
```

## 5. Select optimal number of components

```python
optimal_components = np.argmax(np.cumsum(pca.explained_variance_ratio_) >= 0.95) + 1
pca = PCA(n_components=optimal_components)
X_pca = pca.fit_transform(X)
```

## 6. Save new dataset pca

```python
pd.DataFrame(X_pca).to_csv('../data/heart_disease_pca.csv', index=False)
```
